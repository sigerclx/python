```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]
print(ys)
print(x)
plt.plot(x, ys, '+')
plt.fill_between(x, ys,197,  facecolor='r', alpha=0.3)

plt.title("Sample Visualization")
plt.show()
```
